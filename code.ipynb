## **Load dataset**



```python
import seaborn as sns
iris_df = sns.load_dataset('iris')
```

## **Inspect the data**

Print the shape, column names, and the first few rows of the dataset.


```python
print(f"Shape of the dataset: {iris_df.shape}")
print(f"Column names: {iris_df.columns.tolist()}")
display(iris_df.head())
```

# **Summarize** the data using `.info()` and `.describe()` to get summary statistics.


```python
iris_df.info()
iris_df.describe()
```

## **Data visualizer**

Create a scatter plot, histograms, and box plots to visualize the Iris dataset.


# Scatter plot to visualize the relationship between sepal length and sepal width, color-coded by species.



```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.scatterplot(data=iris_df, x='sepal_length', y='sepal_width', hue='species')
plt.title('Sepal Length vs. Sepal Width')
plt.show()
```

##Histograms for each of the four numerical features to visualize their distributions. A 2x2 subplot grid is appropriate for this.



```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
iris_df.hist(ax=axes)
fig.suptitle('Distribution of Iris Features', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

# Box plots for each numerical feature, grouped by species, to identify potential outliers and compare distributions across species.



```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Box Plots of Iris Features by Species', fontsize=16)

sns.boxplot(ax=axes[0, 0], data=iris_df, x='species', y='sepal_length')
axes[0, 0].set_title('Sepal Length')

sns.boxplot(ax=axes[0, 1], data=iris_df, x='species', y='sepal_width')
axes[0, 1].set_title('Sepal Width')

sns.boxplot(ax=axes[1, 0], data=iris_df, x='species', y='petal_length')
axes[1, 0].set_title('Petal Length')

sns.boxplot(ax=axes[1, 1], data=iris_df, x='species', y='petal_width')
axes[1, 1].set_title('Petal Width')

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

### **Data Analysis Key Findings**

*   The dataset contains 150 entries and 5 columns: `sepal_length`, `sepal_width`, `petal_length`, `petal_width`, and `species`.
*   There are no missing values in the dataset.
*   The numerical columns (`sepal_length`, `sepal_width`, `petal_length`, `petal_width`) have appropriate `float64` data types, and the `species` column has an `object` data type.
*   The scatter plot of sepal length versus sepal width shows distinct groupings by species, with Setosa clearly separated from Versicolor and Virginica.
*   Histograms reveal that `petal_length` and `petal_width` have strongly bimodal distributions, indicating clear separation based on species, while `sepal_width` is roughly normally distributed and `sepal_length` appears somewhat bimodal.
*   Box plots show that Setosa has significantly smaller petal length and petal width compared to Versicolor and Virginica. Virginica generally exhibits the largest measurements across most features.
*   Potential outliers were observed in the `sepal_width` for the Virginica and Setosa species according to the box plots.
